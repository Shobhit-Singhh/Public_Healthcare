```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cardio_train.csv',delimiter=';')
```

```python
df['age']=df['age']/365
```

```python
df
```

```python
df_contnuous=df.drop(['cholesterol','gluc','smoke','alco','active','cardio','id','gender'],axis=1)
```

# Multivariate Normal Distribution

```python
original_data = df_contnuous.to_numpy()

# Calculate the mean vector (μ) of each column
mean_vector = np.mean(original_data, axis=0)

# Calculate the covariance matrix (Σ) that represents the correlations between all the columns
covariance_matrix = np.cov(original_data, rowvar=False)
seed_value = 10

# Generate new data points with the same correlation structure using multivariate normal distribution
num_samples = 30000000  # Number of new data points you want to generate
new_data = np.random.multivariate_normal(mean_vector, covariance_matrix, size=num_samples)

# Convert the generated numpy array back to a pandas DataFrame (if needed)
new_df = pd.DataFrame(new_data, columns=df_contnuous.columns)
```

```python
rounded_columns = ['age', 'height', 'weight', 'ap_hi', 'ap_lo']
new_df[rounded_columns] = new_df[rounded_columns].round(0)
```

```python
new_df=new_df[new_df['age']>15]
new_df=new_df[new_df['height']>120]
new_df=new_df[new_df['weight']>40]
new_df=new_df[new_df['ap_hi']>80]
new_df=new_df[new_df['ap_lo']>60]
```

```python
new_df
```

# Logistic regression Model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Split

```python
X=df.drop(['cholesterol','gluc','smoke','alco','active','cardio','id','gender'],axis=1)
y1=df['cholesterol']
y2=df['gluc']
y3=df['smoke']
y4=df['alco']
y5=df['active']
y6=df['cardio']
y7=df['gender']
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=10)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y2, test_size=0.2, random_state=10)
X_train3, X_test3, y_train3, y_test3 = train_test_split(X, y3, test_size=0.2, random_state=10)
X_train4, X_test4, y_train4, y_test4 = train_test_split(X, y4, test_size=0.2, random_state=10)
X_train5, X_test5, y_train5, y_test5 = train_test_split(X, y5, test_size=0.2, random_state=10)
X_train6, X_test6, y_train6, y_test6 = train_test_split(X, y6, test_size=0.2, random_state=10)
X_train7, X_test7, y_train7, y_test7 = train_test_split(X, y7, test_size=0.2, random_state=10)
```

# Model

```python
model1 = LogisticRegression()
model1.fit(X_train1, y_train1)

model2 = LogisticRegression()
model2.fit(X_train2, y_train2)

model3 = LogisticRegression()
model3.fit(X_train3, y_train3)

model4 = LogisticRegression()
model4.fit(X_train4, y_train4)
          
model5 = LogisticRegression()
model5.fit(X_train5, y_train5)

model6 = LogisticRegression()
model6.fit(X_train6, y_train6)

model7 = LogisticRegression()
model7.fit(X_train7, y_train7)
```

# Prediction

```python
y_pred1 = model1.predict(X_test1)
y_pred2 = model2.predict(X_test2)
y_pred3 = model3.predict(X_test3)
y_pred4 = model4.predict(X_test4)
y_pred5 = model5.predict(X_test5)
y_pred6 = model6.predict(X_test6)
y_pred7 = model7.predict(X_test7)
```

# Scores for all columns

```python
print("Accuracy for cholesterol :", accuracy_score(y_test1, y_pred1))
print("Accuracy for gluc        :", accuracy_score(y_test2, y_pred2))
print("Accuracy for smoke       :", accuracy_score(y_test3, y_pred3))
print("Accuracy for alco        :", accuracy_score(y_test4, y_pred4))
print("Accuracy for active      :", accuracy_score(y_test5, y_pred5))
print("Accuracy for cardio      :", accuracy_score(y_test6, y_pred6))
print("Accuracy for gender      :", accuracy_score(y_test7, y_pred7))
```
