### To handle this Situation:-

## Normalisation male_fraction & female_count column of every district

Lets assume total male_fraction & female_count of Nagaon district(with highest population) as 1

then we can calculate the normalized population of other district

```python
district['normalized_male_count']=district['male_count']/district['male_count'].max()
district['normalized_female_count']=district['female_count']/district['female_count'].max()
```

```python
district
```

### Now if we multiply normalized_male_count with (Male assigned = 24470)

we will get the total number of male assigned to a perticular district 

### Now if we multiply normalized_female_count with (Male assigned = 23540)

we will get the total number of female assigned to a perticular district 

by this way we are not disturbing the gender ratio as well as relative population density of any district

this way we can also preserve the Correlation inbetween the columns also
